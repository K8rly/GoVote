```python
# 1. Create a DataFrame for the Generational Groups data. 
file_to_open = (r"C:\Users\emadn\Class\GoVote\Generational_Groups_df.csv")
Generational_groups_df = pd.read_csv(file_to_open, encoding="ISO-8859-1") 
Generational_groups_df
```

```python
Generational_groups_df.columns
```

```python
Generational_groups_df = Generational_groups_df.drop(Generational_groups_df.columns[[0, 1, 2, 3, 4, 5]], axis=1)
Generational_groups_df.head()
```

```python
#Replace True and False Values with 0 and 1
Generational_groups_df.replace(('D', 'R', 'N', 'L', 'G', 'X', 't', 'f'), (1, 1, 1, 1, 1, 1, 1, 0), inplace=True)
Generational_groups_df
```

```python
Generational_groups_df.to_csv('Generational_GroupS_clean.csv')
```
