```python
# Assign the data to X and y
y=vote_gen_merged["G_11_03_2020"]
X=vote_gen_merged.drop(columns=['city', 'ward', 'pct', 'party','Source','scd',
       'CONG', 'HSE', 'SEN', 'JUD', 'CCD', 'G_11_02_2021', 'G_11_03_2020',
       'G_11_05_2019', 'G_11_06_2018', 'G_11_07_2017', 'G_11_08_2016',
       'G_11_03_2015', 'G_11_04_2014', 'G_11_05_2013', 'G_11_06_2012',
       'G_11_08_2011', 'G_11_02_2010', 'G_11_03_2009', 'G_11_04_2008',
       'G_11_06_2007', 'G_11_07_2006'])
print("Shape: ", X.shape, y.shape)
```

```python
# Split the data by using train_test_split()
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
