```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Create a DataFrame for the Generational Groups data. 
file_to_open = (r"C:\Users\emadn\Class\GoVote\Generational_Groups_df.csv")
Generational_groups_df = pd.read_csv(file_to_open, encoding="ISO-8859-1") 
Generational_groups_df
```

```python
Generational_groups_df.columns
```

```python
Generational_groups_df = Generational_groups_df.drop(Generational_groups_df.columns[[0, 2, 3, 4, 5]], axis=1)
Generational_groups_df.head()
```

```python
#Replace True and False Values with 0 and 1
Generational_groups_df.replace(('D', 'R', 'N', 'L', 'G', 'X', 't', 'f'), (1, 1, 1, 1, 1, 1, 1, 0), inplace=True)
Generational_groups_df
```

```python
Generational_groups_df.fillna(0)
```

```python
# 1. Create a DataFrame for the Registration Dates data. 
file_to_open = (r"C:\Users\emadn\Class\GoVote\Resources\reg_df.csv")
Registration_Dates_df = pd.read_csv(file_to_open, encoding="ISO-8859-1") 
Registration_Dates_df
```

```python
Registration_Dates_df.rename(columns = {'voter_id_org':'Voter ID'}, inplace = True)
Registration_Dates_df
```

```python
# Combine the data into a single dataset.
Reg_Date_df = pd.merge(Generational_groups_df, Registration_Dates_df, on=["Voter ID", "Voter ID"])
Reg_Date_df.head()
```

```python
Reg_Date_df.columns
```

```python
# Reorder the columns
new_column_order = ['Voter ID','Registration Date', '2020 Primary Election', '2016 Primary Election',
       '2012 Primary Election', '2008 Primary Election',
       '2020 General Election', '2016 General Election',
       '2012 General Election', '2008 General Election', 'Generational_Group']
Reg_Date_df= Reg_Date_df[new_column_order]
Reg_Date_df
```
