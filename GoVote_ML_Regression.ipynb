```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Create a DataFrame for the voting data. 
file_to_open = "rvht.csv"
vote_df = pd.read_csv(file_to_open, encoding="ISO-8859-1") 
```

```python
primary_general_df =vote_df[vote_df.columns.drop(list(vote_df.filter(regex='S_|SG_|SP_')))]
primary_general_df.shape
```

```python
# Make dataFrame of Voter Info
voter_df = primary_general_df[['voter_id_org', 'city', 'party','birth_date','zip', 'Registration Date', 'Last Registration Date']]
voter_df.head()
```

```python
# voter_df = voter_df.set_index('voter_id_org')
# voter_df.head()
```

```python
#Make all the zip codes 5 digits instead of combination of 5 and 9 digits
voter_df['zip'] = voter_df['zip'].str[:5]
voter_df.head()
```

```python
#Make Election DataFrame
election_df = primary_general_df[['voter_id_org', 'G_11_03_2020', 'P_03_17_2020', 'G_11_08_2016', 'P_03_15_2016']]
election_df
```

```python
# election_df = election_df.set_index('voter_id_org')
```

```python
election_cleaned = election_df.fillna(0)
```

```python
election_cleaned.head()
```

```python
election_cleaned.replace(
        to_replace="Y",
        value=1,
        inplace=True)
```

```python
one_hot_encoded_data = pd.get_dummies(election_cleaned, columns = ['P_03_17_2020', 'P_03_15_2016'])
one_hot_encoded_data
```
