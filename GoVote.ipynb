```python
import pandas as pd
```

```python
# Files to load
voting_info = ("Resources/rvht.csv")
```

```python
# Read voting file in Pandas DataFrame.
voting_info_df = pd.read_csv(voting_info, low_memory=False)
voting_info_df
```

```python
print(list(voting_info_df.columns))
```

```python
pd.set_option('display.max_rows', None)
voting_info_df.dtypes
```

```python
round(voting_info_df.isna().mean() * 100, 2)
```

```python
voting_info_df = voting_info_df.fillna(0)
```

```python
voting_info_df.replace(('Y', 'D', 'R', 'N', 'X', "I"), (1, 1, 1, 1, 1, 1), inplace=True)
```

```python
general_df =voting_info_df[voting_info_df.columns.drop(list(voting_info_df.filter(regex='S_|SG_|SP_')))]
general_df.shape
```

```python
general_df.to_csv('general_df.csv')
```
