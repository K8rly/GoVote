```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Create a DataFrame for the voting data. 
file_to_open = "rvht.csv"
vote_df = pd.read_csv(file_to_open, encoding="ISO-8859-1") 
```

```python
vote_df['Precint'] = vote_df.agg(lambda x: f"{x['city']}-0{x['ward']}-{x['pct']}", axis=1)
vote_df
```
